## Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
import io
import numpy as np
```

## Import csv file

```python
# dialog to upload from computer
uploaded = files.upload()
file_name = next(iter(uploaded))
table_1 = pd.read_csv(io.BytesIO(uploaded[file_name]))
```

## Store Column names as a list

```python
col_names = table_1.columns
print("\nColumn names:")
for a in range(len(col_names)):
  print(a, " ", col_names[a])
```

```python
# Create a scatterplot of sepal length vs. sepal width

# This dataset did not have pre-assigned column names, so column names needed to be set

# Assign column names
table_1.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

# Get a list of unique species
unique_species = table_1['species'].unique()
species_list = []

for species in unique_species:
    species_list.append(species)

# Filter a plot a scatterplot graph for each unique species
for species in species_list:
  # Filter for current unique species
  data = table_1[table_1['species'] == species]

  # Create x_table and y_table
  x_table = data['sepal_length'].values
  y_table = data['sepal_width'].values
  plt.scatter(x_table, y_table)
  plt.xlabel('Sepal Length')
  plt.ylabel('Sepal Width')
  plt.title('Iris Dataset: Sepal Length vs. Sepal Width: ' + species)
  plt.show()
  print(" ")
```
